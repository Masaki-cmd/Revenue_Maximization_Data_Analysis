## 1. 課題定義

まず、提供された各データセットを読み込み、その構造と内容を把握します。これにより、データからどのような情報を引き出せるか、またどのようなビジネス上の課題を特定できるかの見当をつけます。

````python
import pandas as pd

# customers.csv を読み込む
customers_df = pd.read_csv('/content/customers.csv')
print('--- customers.csv ---')
display(customers_df.head())
print(customers_df.info())
````

### 1.1. データクレンジングと前処理

データの正確な分析のためには、まずデータ型を適切に変換し、欠損値や異常値がないかを確認し、必要に応じて処理する必要があります。

#### `customers_df` のクリーンアップ

`registration_date` が `object` 型になっているため、`datetime` 型に変換します。これにより、時間軸での分析や顧客の登録期間の計算などが可能になります。

````python
# registration_date を datetime 型に変換
customers_df['registration_date'] = pd.to_datetime(customers_df['registration_date'])

print('--- customers_df (データ型変換後) ---')
print(customers_df.info())
display(customers_df.head())
````

#### `orders_df` のクリーンアップ

`order_date` と `delivery_date` が `object` 型になっているため、`datetime` 型に変換します。また、`customer_rating` に欠損値が多く含まれているため、どのように処理するか検討が必要です。今回は、欠損値がある場合は評価がされていないとみなし、分析によっては除外するか、中央値などで補完することを検討します。ここでは、評価がないことを示すためにNaNをそのまま保持します。

````python
# order_date と delivery_date を datetime 型に変換
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])
orders_df['delivery_date'] = pd.to_datetime(orders_df['delivery_date'])

print('\n--- orders_df (データ型変換後) ---')
print(orders_df.info())
display(orders_df.head())

# customer_rating の欠損値を確認
print(f"\ncustomer_rating の欠損値数: {orders_df['customer_rating'].isnull().sum()}")
````

#### データの統合

顧客ごとの購買行動と属性を合わせて分析するために、`orders_df`と`customers_df`を`customer_id`をキーとして結合します。これにより、個々の注文がどのような顧客によって行われたのかを詳細に把握できるようになります。

**なぜこのグラフを作ったのか**: 顧客の属性（年齢、性別、会員ティアなど）が購入行動（購入商品、購入金額、頻度など）にどのように影響するかを分析するための基盤を構築するためです。例えば、特定の年齢層が好む商品カテゴリや、会員ティアが高い顧客の平均注文額などを調べることができます。

````python
# orders_df と customers_df を customer_id で結合
merged_df = pd.merge(orders_df, customers_df, on='customer_id', how='left')

print('\n--- 結合後のデータフレーム (merged_df) ---')
display(merged_df.head())
print(merged_df.info())
````

## 2. 仮説設定とデータ分析の方向性

データのクレンジングと統合が完了しました。ここからは、具体的な分析を進め、収益最大化のための課題を特定し、仮説を立てていきます。

まず、全体の収益トレンドを把握するために、`monthly_revenue_df` を使って時系列での収益の推移を確認します。これにより、季節性や成長傾向、異常値などを発見し、改善の機会を見つけることができます。

**なぜこの分析をするのか**: ビジネスの健全性を評価し、収益の増減に影響を与える要因を特定する第一歩として、マクロな視点から収益の動向を理解することが不可欠です。これにより、次にどの領域（顧客、製品など）に焦点を当てるべきかの方針を立てることができます。

### 2.1. 製品パフォーマンスの分析

月次収益の全体像を把握したところで、次に製品レベルに焦点を当てます。`product_summary_df` を使用して、どの製品カテゴリや個々の製品が収益に大きく貢献しているのか、または改善の余地があるのかを特定します。

**なぜこの分析をするのか**: 収益の源泉となっている製品を理解し、その成功要因を他の製品に応用するためです。また、収益が低い、返品率が高い、割引率が高いなどの問題を抱える製品を特定し、改善策を検討するための基礎情報となります。これにより、製品ポートフォリオの最適化やプロモーション戦略の改善に繋がる仮説を立てることができます。

````python
# カテゴリごとの総収益と総注文数を集計
category_revenue = product_summary_df.groupby('category').agg(
    total_revenue_usd=('total_revenue_usd', 'sum'),
    total_orders=('total_orders', 'sum')
).reset_index().sort_values(by='total_revenue_usd', ascending=False)

print('\n--- カテゴリ別収益と注文数 ---')
display(category_revenue.head(10))

# 上位Nカテゴリの収益を可視化
plt.figure(figsize=(12, 7))
sns.barplot(x='total_revenue_usd', y='category', data=category_revenue.head(10), palette='viridis')
plt.title('カテゴリ別総収益トップ10')
plt.xlabel('総収益 (USD)')
plt.ylabel('カテゴリ')
plt.grid(axis='x')
plt.tight_layout()
plt.show()

# 返品率の高い製品カテゴリの特定
category_return_rate = product_summary_df.groupby('category')['return_rate'].mean().reset_index().sort_values(by='return_rate', ascending=False)

print('\n--- カテゴリ別平均返品率 ---')
display(category_return_rate.head(10))

plt.figure(figsize=(12, 7))
sns.barplot(x='return_rate', y='category', data=category_return_rate.head(10), palette='magma')
plt.title('カテゴリ別平均返品率トップ10')
plt.xlabel('平均返品率 (%)')
plt.ylabel('カテゴリ')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
````

### 2.2. 顧客行動の分析 (続き)

#### 獲得チャネル別分析

顧客がどのような経路（獲得チャネル）でサービスにたどり着いたかが、その後の購買行動に影響を与える可能性があります。獲得チャネル別に総支出額や注文数を比較することで、効果的なマーケティングチャネルを特定し、リソース配分を最適化するヒントを得ます。

**なぜこの分析をするのか**: 効率的な顧客獲得戦略を立てるためです。高収益顧客をもたらすチャネルに投資を集中したり、改善が必要なチャネルを特定して最適化したりすることで、マーケティングROI（投資収益率）を向上させることができます。

````python
# 獲得チャネル別の平均総支出額を計算
channel_spend = merged_df.groupby('acquisition_channel')['total_spend_usd'].mean().reset_index().sort_values(by='total_spend_usd', ascending=False)

print('\n--- 獲得チャネル別平均総支出額 ---')
display(channel_spend)

plt.figure(figsize=(12, 7))
sns.barplot(x='acquisition_channel', y='total_spend_usd', data=channel_spend, palette='viridis')
plt.title('獲得チャネル別平均総支出額')
plt.xlabel('獲得チャネル')
plt.ylabel('平均総支出額 (USD)')
plt.grid(axis='y')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 獲得チャネル別の平均注文数を計算
channel_orders = merged_df.groupby('acquisition_channel')['order_id'].nunique().reset_index()
channel_customers = merged_df.groupby('acquisition_channel')['customer_id'].nunique().reset_index()
channel_avg_orders = pd.merge(channel_orders, channel_customers, on='acquisition_channel')
channel_avg_orders['avg_orders_per_customer'] = channel_avg_orders['order_id'] / channel_avg_orders['customer_id']
channel_avg_orders = channel_avg_orders.sort_values(by='avg_orders_per_customer', ascending=False)

print('\n--- 獲得チャネル別顧客あたりの平均注文数 ---')
display(channel_avg_orders)

plt.figure(figsize=(12, 7))
sns.barplot(x='acquisition_channel', y='avg_orders_per_customer', data=channel_avg_orders, palette='magma')
plt.title('獲得チャネル別顧客あたりの平均注文数')
plt.xlabel('獲得チャネル')
plt.ylabel('顧客あたりの平均注文数')
plt.grid(axis='y')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

#### 課題の特定と仮説構築

上記の分析結果から、以下のような課題と仮説が考えられます。

**課題**: 特定のカテゴリが高収益を上げている一方で、平均返品率が高いカテゴリも存在する。

**仮説1 (収益最大化)**: 高収益カテゴリの製品ラインナップを強化したり、関連商品をバンドル販売することで、さらなる収益増加が見込めるのではないか。

**仮説2 (コスト最適化)**: 返品率が高いカテゴリの製品について、製品品質の見直し、商品説明の改善、または顧客サポートの強化により、返品に伴うコストを削減できるのではないか。

次のステップでは、これらの仮説をさらに深掘りするために、顧客行動データ (`merged_df`) を用いて、顧客セグメントごとの特徴や購買パターンを分析します。

### 2.2. 顧客行動の分析

ここからは、`merged_df` を使用して、顧客の属性と購買行動の関連性を分析します。これにより、どのような顧客が収益に貢献しているのか、または改善の余地がある顧客セグメントはどこかといった具体的な洞察を得ることができます。

**なぜこの分析をするのか**: 顧客セグメントごとに異なるマーケティング戦略やプロモーションを立案するためです。例えば、高価値顧客の特性を理解し、同様の特性を持つ新規顧客獲得に注力したり、離反リスクのある顧客に再エンゲージメントの施策を打つことができます。


#### 顧客の年齢層別分析

顧客の年齢層が総支出額にどのように影響するかを可視化します。これにより、特定の年齢層がより多くの支出をしているか、あるいは特定の年齢層にリーチできていないかなどを把握できます。

**なぜこのグラフを作ったのか**: ターゲット顧客層の特定と、年齢層に合わせた製品開発やマーケティングメッセージの調整のヒントを得るためです。例えば、特定の年齢層が特定のカテゴリに興味を示している場合、そのカテゴリのプロモーションを強化することができます。

````python
# 年齢層を定義
bins = [10, 20, 30, 40, 50, 60, 70, 80]
labels = ['10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
merged_df['age_group'] = pd.cut(merged_df['age'], bins=bins, labels=labels, right=False)

# 年齢層別の平均総支出額を計算
age_spend = merged_df.groupby('age_group')['total_spend_usd'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='total_spend_usd', data=age_spend, palette='crest')
plt.title('年齢層別平均総支出額')
plt.xlabel('年齢層')
plt.ylabel('平均総支出額 (USD)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
````

#### 会員ティア別分析

顧客の会員ティア（Free, Silver, Gold, Platinum）が総支出額や平均注文額にどのように影響するかを分析します。これにより、上位ティア顧客の価値を再認識し、会員プログラムの改善やアップセル戦略のヒントを得ることができます。

**なぜこのグラフを作ったのか**: 会員ティアの価値を評価し、より多くの顧客を上位ティアに引き上げるための戦略を検討するためです。例えば、GoldやPlatinum会員の購買行動を分析し、Free会員やSilver会員をこれらの高価値顧客に育成するための施策を考案できます。

````python
# 会員ティア別の平均総支出額と平均注文額を計算
membership_spend = merged_df.groupby('membership_tier').agg(
    avg_total_spend=('total_spend_usd', 'mean'),
    avg_order_value=('avg_order_value_usd', 'mean')
).reset_index().sort_values(by='avg_total_spend', ascending=False)

print('\n--- 会員ティア別平均支出額 ---')
display(membership_spend)

# 可視化
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.barplot(x='membership_tier', y='avg_total_spend', data=membership_spend, ax=axes[0], palette='magma', order=['Free', 'Silver', 'Gold', 'Platinum'])
axes[0].set_title('会員ティア別平均総支出額')
axes[0].set_xlabel('会員ティア')
axes[0].set_ylabel('平均総支出額 (USD)')
axes[0].grid(axis='y')

sns.barplot(x='membership_tier', y='avg_order_value', data=membership_spend, ax=axes[1], palette='magma', order=['Free', 'Silver', 'Gold', 'Platinum'])
axes[1].set_title('会員ティア別平均注文額')
axes[1].set_xlabel('会員ティア')
axes[1].set_ylabel('平均注文額 (USD)')
axes[1].grid(axis='y')

plt.tight_layout()
plt.show()
````

#### 獲得チャネル別分析

顧客がどのような経路（獲得チャネル）でサービスにたどり着いたかが、その後の購買行動に影響を与える可能性があります。獲得チャネル別に総支出額や注文数を比較することで、効果的なマーケティングチャネルを特定し、リソース配分を最適化するヒントを得ます。

**なぜこの分析をするのか**: 効率的な顧客獲得戦略を立てるためです。高収益顧客をもたらすチャネルに投資を集中したり、改善が必要なチャネルを特定して最適化したりすることで、マーケティングROI（投資収益率）を向上させることができます。

````python
# 獲得チャネル別の平均総支出額を計算
channel_spend = merged_df.groupby('acquisition_channel')['total_spend_usd'].mean().reset_index().sort_values(by='total_spend_usd', ascending=False)

print('\n--- 獲得チャネル別平均総支出額 ---')
display(channel_spend)

plt.figure(figsize=(12, 7))
sns.barplot(x='acquisition_channel', y='total_spend_usd', data=channel_spend, palette='viridis')
plt.title('獲得チャネル別平均総支出額')
plt.xlabel('獲得チャネル')
plt.ylabel('平均総支出額 (USD)')
plt.grid(axis='y')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 獲得チャネル別の平均注文数を計算
channel_orders = merged_df.groupby('acquisition_channel')['order_id'].nunique().reset_index()
channel_customers = merged_df.groupby('acquisition_channel')['customer_id'].nunique().reset_index()
channel_avg_orders = pd.merge(channel_orders, channel_customers, on='acquisition_channel')
channel_avg_orders['avg_orders_per_customer'] = channel_avg_orders['order_id'] / channel_avg_orders['customer_id']
channel_avg_orders = channel_avg_orders.sort_values(by='avg_orders_per_customer', ascending=False)

print('\n--- 獲得チャネル別顧客あたりの平均注文数 ---')
display(channel_avg_orders)

plt.figure(figsize=(12, 7))
sns.barplot(x='acquisition_channel', y='avg_orders_per_customer', data=channel_avg_orders, palette='magma')
plt.title('獲得チャネル別顧客あたりの平均注文数')
plt.xlabel('獲得チャネル')
plt.ylabel('顧客あたりの平均注文数')
plt.grid(axis='y')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

#### 課題の特定と仮説構築 (顧客行動)

年齢層別、会員ティア別、獲得チャネル別の分析から、以下の課題と仮説が考えられます。

**課題1**: 特定の年齢層（例: 20代、30代）が高い支出傾向を示しているが、60代以上では支出が減少傾向にある。また、会員ティアが高い顧客ほど平均支出額・平均注文額が大きい。

**仮説1 (収益最大化)**:
*   高支出の年齢層（20代、30代）向けに、さらにパーソナライズされたプロモーションや新製品の提案を行うことで、顧客単価の向上とリピート購入を促進できるのではないか。
*   高価値のGold/Platinum会員層へのアップセル、またはFree/Silver会員を上位ティアに引き上げるためのインセンティブプログラム（例: 限定クーポン、早期アクセス、ポイント還元率向上）を導入することで、全体の収益を増加できるのではないか。
*   支出が少ない年齢層（60代以上）向けに、ニーズに合った製品カテゴリ（例: 健康関連商品、生活支援サービス）の拡充や、使いやすいUI/UXの改善を行うことで、新たな顧客層を開拓できるのではないか。

**課題2**: 獲得チャネルによって顧客の平均支出額や顧客あたりの平均注文数に差がある。

**仮説2 (コスト最適化・収益最大化)**:
*   平均支出額や顧客あたりの平均注文数が高い獲得チャネル（例: Paid Ad, Direct, Referral）へのマーケティング投資を増やすことで、より効率的な顧客獲得と収益増加に繋がるのではないか。
*   平均支出額が低いチャネル（例: Social Media）については、獲得後の顧客育成プログラムを強化したり、製品のレコメンデーションを最適化したりすることで、LTV（顧客生涯価値）を向上できるのではないか。

## 3. 総合的な洞察と提案

ここまでのデータ分析（月次収益トレンド、製品パフォーマンス、顧客行動）から得られた主要な洞察をまとめ、収益の最大化およびコストの最適化に向けた具体的な提案を行います。

### 3.1. 主要な洞察

1.  **月次収益の安定性と成長機会**: 月次収益は比較的安定しているものの、大幅な成長トレンドは見られない。特定の季節性やイベントによる収益変動は限定的である。
2.  **製品カテゴリの収益貢献と返品率**: `Electronics`が圧倒的な収益を上げている一方で、`Travel & Luggage`や`Pet Supplies`など、一部カテゴリでは返品率が高い傾向にある。高収益カテゴリの成功要因を深掘りし、高返品率カテゴリの改善が必要。
3.  **顧客の価値と行動パターン**:
    *   **年齢層**: 20-39歳が高支出傾向にあり、60代以上の支出は控えめである。
    *   **会員ティア**: PlatinumやGold会員は平均支出額および平均注文額がFree会員と比較して大幅に高く、非常に価値の高い顧客セグメントであることが確認された。
    *   **獲得チャネル**: Paid Ad、Direct、Referralが平均総支出額が高い顧客を獲得しており、効率的なチャネルである可能性が高い。一方で、Social Mediaからの顧客は平均支出額が低い傾向にある。

### 3.2. 収益最大化のための提案

1.  **高価値顧客層 (Platinum/Gold会員、20-39歳) へのエンゲージメント強化**
    *   **特別プロモーション**: 既存のPlatinum/Gold会員向けに、限定先行販売、VIPカスタマーサービス、パーソナライズされた割引を提供し、ロイヤリティをさらに高める。
    *   **アップセル・クロスセル**: Free/Silver会員に対して、購入履歴や閲覧行動に基づいた上位ティアへのアップグレードインセンティブ（例: 一定額以上の購入でGold会員へ自動アップグレード）や、関連商品の推奨を強化する。
    *   **ターゲット広告**: 20-39歳の高支出層が頻繁に利用するプラットフォーム（SNS、特定のオンラインメディア）での広告出稿を強化し、ニーズに合った製品を効果的にアピールする。

2.  **主力製品カテゴリ (`Electronics`, `Clothing & Apparel`, `Home & Kitchen`) の戦略的強化**
    *   **新製品開発・拡充**: これらの高収益カテゴリにおいて、市場トレンドを捉えた新製品を積極的に投入し、顧客の関心を維持する。
    *   **バンドル販売**: 関連性の高い製品をセットで割引販売することで、顧客単価の向上を狙う。
    *   **在庫最適化**: 需要の高い製品の欠品を防ぎ、販売機会損失を最小限に抑える。

3.  **効果的な獲得チャネルへの投資集中と最適化**
    *   **広告予算の再配分**: Paid Ad、Direct、Referralといった高効率チャネルへの広告予算配分を増やし、これらのチャネルからの新規顧客獲得を加速させる。
    *   **Referralプログラムの強化**: 既存顧客が新規顧客を紹介した際に、双方にメリットのあるインセンティブ（例: 次回購入割引、ポイント付与）を提供し、Referralチャネルからの顧客増加を図る。

4.  **顧客エンゲージメントの向上**
    *   **パーソナライズされたレコメンデーション**: 顧客の購買履歴、閲覧履歴、登録カテゴリに基づいて、個々に最適化された製品レコメンデーションをメール、アプリ通知、ウェブサイト上で提供する。
    *   **ウィッシュリスト活用**: ウィッシュリストに登録された商品の価格変動通知や、在庫状況の更新通知を行い、購入を促す。

### 3.3. コスト最適化のための提案

1.  **高返品率カテゴリ (`Travel & Luggage`, `Pet Supplies`) の改善**
    *   **製品品質の向上**: 返品理由を詳細に分析し、品質問題が原因であればサプライヤーとの連携を強化し、製品改善を促す。
    *   **商品説明の明確化**: 製品画像、仕様、サイズ情報などをより正確かつ詳細に提供することで、購入時の期待値と実物のギャップを解消し、返品率を低減する。
    *   **レビューの活用**: 返品が多い製品については、既存顧客からのレビューを積極的に収集し、製品ページに目立つように表示することで、新規顧客の意思決定を支援する。

2.  **割引戦略の最適化**
    *   **ターゲット割引**: 一律の割引ではなく、特定の顧客セグメント（例: 新規顧客、休眠顧客、特定のカテゴリの購入者）に限定した割引や、購入金額に応じた段階的な割引を導入する。
    *   **割引率のモニタリング**: 割引率が高いにもかかわらず販売促進効果が薄い製品やカテゴリがないか継続的にモニタリングし、不要な割引を削減する。

3.  **低効率獲得チャネル (`Social Media`) の見直し**
    *   **獲得後のLTV向上施策**: Social Media経由で獲得した顧客のエンゲージメントが低い場合、獲得後のフォローアップ施策（例: 魅力的なコンテンツ配信、限定イベントへの招待）を強化し、長期的な顧客価値を高める。
    *   **チャネル戦略の再評価**: 費用対効果が低いと判断される場合、Social Media広告のクリエイティブやターゲット設定を見直すか、予算を他の高効率チャネルに再配分することを検討する。

### 2.3. RFM分析による顧客セグメンテーション

これまでの分析で顧客行動の全体像を把握しましたが、個々の顧客の価値をより深く理解するためには、RFM（Recency、Frequency、Monetary）分析が非常に有効です。

**Recency**: 顧客が最後に購入してからの経過日数。短いほど優良顧客。
**Frequency**: 顧客の購入頻度。高いほど優良顧客。
**Monetary**: 顧客の購入金額。高いほど優良顧客。

これらの指標を用いて顧客をセグメンテーションすることで、各セグメントに最適化されたマーケティング戦略を立案できます。

**なぜこの分析をするのか**: 顧客をその行動に基づいて明確に分類することで、高価値顧客の維持、休眠顧客の再活性化、新規顧客の育成といった、よりパーソナライズされた顧客戦略を効果的に実施するためです。

````python
# RFM分析のために基準日を設定（データセット内の最新購入日より1日後）
current_date = merged_df['order_date'].max() + pd.Timedelta(days=1)

# customer_id ごとに Recency, Frequency, Monetary を計算
rfm_df = merged_df.groupby('customer_id').agg(
    Recency=('order_date', lambda date: (current_date - date.max()).days), # 最終購入からの日数
    Frequency=('order_id', 'nunique'), # ユニークな注文数
    Monetary=('total_amount_usd', 'sum') # 総支出額
).reset_index()

print('--- RFMデータフレーム ---')
display(rfm_df.head())
print(rfm_df.describe())
````

RFMスコアを計算するために、各RFM指標をランク付けします。ここでは、各指標を5段階（1-5）にスコアリングします。

*   **Recency**: 日数が短いほど高スコア（5が最高）
*   **Frequency**: 頻度が高いほど高スコア（5が最高）
*   **Monetary**: 金額が高いほど高スコア（5が最高）

````python
# Function to calculate RFM scores
def get_rfm_scores(df, column, n_quantiles=5, ascending=True):
    # Use qcut to get quantiles, without assigning labels initially
    # labels=False makes qcut return integer indices for the bins
    qcut_result = pd.qcut(df[column], n_quantiles, labels=False, duplicates='drop')

    # The result is 0-indexed. Add 1 to make it 1-indexed score.
    scores = qcut_result + 1

    # If ascending=False (for Recency), invert the scores
    if not ascending:
        # Get the max score based on the actual number of unique bins created
        max_score = scores.max()
        scores = max_score - scores + 1
    return scores.astype(int)

# RFMスコアの計算
rfm_df['R_Score'] = get_rfm_scores(rfm_df, 'Recency', n_quantiles=5, ascending=False) # Recency: 小さいほど高スコア
rfm_df['F_Score'] = get_rfm_scores(rfm_df, 'Frequency', n_quantiles=5, ascending=True) # Frequency: 大きいほど高スコア
rfm_df['M_Score'] = get_rfm_scores(rfm_df, 'Monetary', n_quantiles=5, ascending=True) # Monetary: 大きいほど高スコア

# RFMセグメントの作成 (RFMスコアの結合)
rfm_df['RFM_Segment'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)

# 合計RFMスコアの計算
rfm_df['RFM_Score'] = rfm_df['R_Score'] + rfm_df['F_Score'] + rfm_df['M_Score']

print('--- RFMスコア付きデータフレーム ---')
display(rfm_df.head())

# RFMスコア分布の確認
plt.figure(figsize=(10, 6))
sns.histplot(rfm_df['RFM_Score'], bins=range(3, 17), kde=True)
plt.title('RFMスコア分布')
plt.xlabel('RFMスコア')
plt.ylabel('顧客数')
plt.grid(axis='y')
plt.show()
````

RFMスコアに基づいて、顧客をいくつかの主要なセグメントに分類します。例えば、以下のようなセグメントが考えられます。

*   **Champions (優良顧客)**: 最新の購入日、高頻度、高金額
*   **Loyal Customers (優良リピーター)**: 高頻度で良い金額の購入
*   **Potential Loyalists (優良予備軍)**: 最近購入したが、頻度と金額は中程度
*   **New Customers (新規顧客)**: 最近購入したが、頻度と金額は低い
*   **At Risk (離反リスク)**: かなり前に購入したが、頻度と金額は中程度
*   **Cant't Lose Them (離反寸前)**: 非常に前に購入し、頻度と金額も低い
*   **Lost (離反顧客)**: 非常に昔に購入し、頻度と金額も低い

````python
# RFMセグメントを定義する関数
def rfm_segment(df):
    if df['RFM_Score'] >= 13:
        return 'Champions'
    elif df['RFM_Score'] >= 10:
        return 'Loyal Customers'
    elif df['RFM_Score'] >= 8:
        return 'Potential Loyalists'
    elif df['RFM_Score'] >= 6:
        return 'New Customers'
    elif df['RFM_Score'] >= 4:
        return 'At Risk'
    else:
        return 'Lost'

rfm_df['Customer_Segment'] = rfm_df.apply(rfm_segment, axis=1)

print('--- 顧客セグメント分布 ---')
display(rfm_df['Customer_Segment'].value_counts().sort_index())

# 顧客セグメントの可視化
plt.figure(figsize=(12, 7))
sns.countplot(x='Customer_Segment', data=rfm_df, palette='Spectral', order=rfm_df['Customer_Segment'].value_counts().sort_index().index)
plt.title('顧客セグメントの分布')
plt.xlabel('顧客セグメント')
plt.ylabel('顧客数')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 各セグメントのRFM指標の平均値
segment_summary = rfm_df.groupby('Customer_Segment').agg(
    Avg_Recency=('Recency', 'mean'),
    Avg_Frequency=('Frequency', 'mean'),
    Avg_Monetary=('Monetary', 'mean'),
    Count=('customer_id', 'count')
).reset_index().sort_values(by='Avg_Monetary', ascending=False)

print('\n--- 各セグメントのRFM指標サマリー ---')
display(segment_summary)
````

### 3.4. RFM分析に基づく提案

RFM分析の結果、顧客は複数のセグメントに分類されました。これらのセグメントごとに異なるアプローチを適用することで、より効果的な収益最大化およびコスト最適化が期待できます。

1.  **Champions (優良顧客) への戦略**:
    *   **VIPプログラムの強化**: 排他的なプロモーション、新製品の先行アクセス、特別なカスタマーサービスを提供し、ロイヤリティをさらに高める。
    *   **限定イベントへの招待**: 顧客満足度とブランドエンゲージメントを高めるための限定イベントやワークショップを開催する。
    *   **フィードバックの収集**: 製品開発やサービス改善のための重要なフィードバック源として活用する。

2.  **Loyal Customers (優良リピーター) への戦略**:
    *   **パーソナライズされた推奨**: 過去の購入履歴に基づいた関連製品やアップセル・クロスセルの提案を強化する。
    *   **ロイヤリティポイントの提供**: 継続的な購入を促すためのポイントプログラムや割引制度を提供する。

3.  **Potential Loyalists (優良予備軍) への戦略**:
    *   **リピート購入の促進**: 初回購入後のフォローアップメールや、次回の購入を促す限定的な割引クーポンを提供する。
    *   **エンゲージメントコンテンツ**: 製品の使用方法、メリット、関連情報などの価値あるコンテンツを提供し、顧客のエンゲージメントを高める。

4.  **New Customers (新規顧客) への戦略**:
    *   **オンボーディングプログラム**: 購入後のスムーズな顧客体験をサポートするための情報提供や、初回限定の特典を提供する。
    *   **ブランド教育**: 製品の魅力やブランドの価値を伝えるコンテンツを提供し、リピーターへの移行を促す。

5.  **At Risk (離反リスク) / Lost (離反顧客) への戦略**:
    *   **再エンゲージメントキャンペーン**: 特別な割引、限定オファー、またはパーソナライズされたメッセージを通じて、顧客の再購入を促す。
    *   **フィードバックの収集**: 離反理由を特定するためのアンケートやインタビューを実施し、サービス改善に役立てる。

これらのRFMセグメントに基づいた戦略を導入することで、顧客のライフサイクル全体にわたって価値を最大化し、収益向上と顧客ロイヤリティの強化を図ることができます。

### 3.5. RFMセグメント別収益最大化のための具体的な施策（詳細版）

RFM分析によって顧客を価値の高い順にセグメント化しました。各セグメントの特性を深く理解し、それぞれに最適化されたマーケティング戦略を展開することで、収益の最大化を図ります。

#### 1. Champions (優良顧客):
**戦略目標**: 顧客満足度を最高レベルに維持し、ブランドアンバサダーとして育成。彼らのLTVをさらに最大化する。
*   **排他的な体験の提供**: 新製品の限定先行アクセス、VIP専用のカスタマーサポートライン、特別なイベントへの招待（例: デザイナーとの交流会、新商品発表会）。
*   **パーソナライズされたギフトと特典**: 誕生日や記念日に合わせた個別メッセージ付きのギフト、購入履歴に基づいた特別な割引コード。
*   **フィードバックの優先的な取り入れ**: 製品開発やサービス改善に関する意見を積極的に求め、彼らの意見が反映されていることを示す。
*   **紹介プログラムの強化**: 彼らが友人・知人を紹介した際に、紹介者・被紹介者双方に非常に魅力的なインセンティブを提供する。

#### 2. Loyal Customers (優良リピーター):
**戦略目標**: 継続的な購買行動を促進し、Championsへと育成する。
*   **継続購入インセンティブ**: 連続購入回数に応じた割引、特定の期間内での複数回購入でボーナスポイント付与。
*   **カテゴリ横断的なレコメンデーション**: 過去の購入履歴から類推される関連カテゴリの製品や、上位グレードの製品を積極的に推奨。
*   **ロイヤリティプログラムの強化**: ポイントの有効期限延長、特定のティア限定セールへの招待、上位ティアへの昇格条件の明確化と動機付け。
*   **コミュニティ形成**: 専用オンライングループやフォーラムへの招待を通じて、ブランドへの帰属意識を高める。

#### 3. Potential Loyalists (優良予備軍):
**戦略目標**: リピート頻度と購入金額を向上させ、Loyal Customersへと昇格させる。
*   **初回リピート購入促進**: 初回購入後、一定期間内に利用できる「次回購入割引クーポン」の送付。
*   **商品知識の深化**: 購入した商品の活用方法、関連商品の紹介、顧客の興味に合わせたコンテンツ配信（メールマガジン、ブログ記事）。
*   **パーソナルショッパーサービス**: 購買意欲が高まっているタイミングで、個別の相談に対応するパーソナルショッパーの利用を促す。
*   **限定セールへの早期アクセス**: 特定のキャンペーンやセール情報を、一般公開より少し早く提供する。

#### 4. New Customers (新規顧客):
**戦略目標**: 初回購入後の離反を防ぎ、リピーターへと育成する。
*   **充実したオンボーディング**: 購入商品の使い方ガイド、よくある質問集、カスタマーサポートへのスムーズな導線を提供。
*   **ブランドの価値訴求**: ブランドの哲学、製品のこだわり、他社との差別化ポイントを明確に伝え、顧客との心理的なつながりを構築。
*   **限定的な割引と特典**: 初回購入者限定の「リピート購入割引」や、特定のカテゴリで使えるクーポンを提供。
*   **パーソナライズされたフォローアップ**: 閲覧履歴や初回購入商品に基づいた関連商品の紹介、カスタマージャーニーの設計。

#### 5. At Risk (離反リスク) / Lost (離反顧客):
**戦略目標**: 顧客の再活性化。低コストでの再エンゲージメントを試みる。
*   **再エンゲージメントキャンペーン**: 長期間購入のない顧客に対し、限定的な期間で大幅な割引や魅力的なセット商品を提供。
*   **フィードバックの収集**: 離反理由に関するアンケートを実施し、顧客の不満点や期待を把握。改善に活かす。
*   **新しい価値の提供**: 新製品や新サービスに関する情報、以前購入した商品のアップグレード版などの情報を提供し、再度の興味を引く。
*   **顧客属性に応じたメッセージング**: 離反した顧客の購入履歴や属性に基づき、個別に最適化されたメッセージでアプローチする。

これらの施策を各セグメントの顧客に対して継続的に実行し、効果測定と改善を繰り返すことで、全体の収益最大化に繋げます。

**なぜこのグラフを作ったのか**: 会員ティアの価値を評価し、より多くの顧客を上位ティアに引き上げるための戦略を検討するためです。例えば、GoldやPlatinum会員の購買行動を分析し、Free会員やSilver会員をこれらの高価値顧客に育成するための施策を考案できます。

````python
# 会員ティア別の平均総支出額と平均注文額を計算
membership_spend = merged_df.groupby('membership_tier').agg(
    avg_total_spend=('total_spend_usd', 'mean'),
    avg_order_value=('avg_order_value_usd', 'mean')
).reset_index().sort_values(by='avg_total_spend', ascending=False)

print('\n--- 会員ティア別平均支出額 ---')
display(membership_spend)

# 可視化
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.barplot(x='membership_tier', y='avg_total_spend', data=membership_spend, ax=axes[0], palette='magma', order=['Free', 'Silver', 'Gold', 'Platinum'])
axes[0].set_title('会員ティア別平均総支出額')
axes[0].set_xlabel('会員ティア')
axes[0].set_ylabel('平均総支出額 (USD)')
axes[0].grid(axis='y')

sns.barplot(x='membership_tier', y='avg_order_value', data=membership_spend, ax=axes[1], palette='magma', order=['Free', 'Silver', 'Gold', 'Platinum'])
axes[1].set_title('会員ティア別平均注文額')
axes[1].set_xlabel('会員ティア')
axes[1].set_ylabel('平均注文額 (USD)')
axes[1].grid(axis='y')

plt.tight_layout()
plt.show()
````
